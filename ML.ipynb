```
import numpy as np
import pandas as pd
import altair as alt
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```
# Cargar dataset
df = pd.read_csv('vgsales.csv')
```

```

#variables categoricas a numericas utilizando One-Hot Encoding
modelo = X = pd.get_dummies(df[['Platform', 'Genre', 'NA_Sales']], drop_first=True)
#LabelEncoder
label_encoder = LabelEncoder()

#variables categóricas como números enteros
df['Plataforma_Label'] = label_encoder.fit_transform(df['Platform'])
df['Género_Label'] = label_encoder.fit_transform(df['Genre'])

#usar las columnas codificadas en vez de las originales
X = df[['Plataforma_Label', 'Género_Label', 'NA_Sales']]

df['Ventas_Altas_Bajas'] = (df['Global_Sales'] > 1).astype(int)

#convertir variables categóricas a numéricas utilizando One-Hot Encoding
X = pd.get_dummies(df[['Platform', 'Genre', 'NA_Sales']], drop_first=True)
y = df['Ventas_Altas_Bajas']

#entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar modelo
modelo = RandomForestClassifier()
modelo.fit(X_train, y_train)

#predecir
y_pred = modelo.predict(X_test)

#evaluar precisión
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy * 100:.2f}%')

#matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

#guarda el modelo 
#joblib.dump(modelo, 'modelo_ventas_EU.pkl')
```
